```python
x = np.linspace(1, 10, 1000)
y1 = np.power(x, 0.5)
y2 = 2 * np.sin(x)
y3 = 3 * np.cos(x)
```

```python
plt.plot(x, y1, label='$\sqrt{x}$')
plt.plot(x, y2, label='$2 \dot \sin{x}$')
plt.plot(x, y3, 'b--', label='$3 \dot \cos{x}$')

plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Comparison of functions')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 2), dpi=300)
ax1, ax2 = axes
ax1.plot(x, y1)
ax2.plot(x, y2)
ax1.set_xlabel('x')
ax1.set_xlabel('f(x)')
fig.suptitle('2 plots')
```

```python
names = ['a', 'b', 'c']
values = [100, 150, 180]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10), dpi=300)

(ax1, ax2), (ax3, ax4) = axes

ax1.bar(names, values, color=['r', 'g', 'b'])
ax1.set_title('bar')
ax2.scatter(names, values)
ax2.set_title('scatter')
ax3.plot(names, values)
ax3.set_title('plot')
ax4.pie(values)
ax4.set_title('pie')
```

```python
x = np.linspace(0,2*np.pi,500)
y1 = np.sin(x) + np.cos(x)
y2 = np.sin(x) * np.cos(x)

plt.plot(x,y1,color = "Red", label = "np.sin(x) + np.cos(x)")
plt.plot(x,y2,color = "Green", label = "np.sin(x) * np.cos(x)")
plt.title("   sin(x), cos(x)")
plt.xlabel("x ()")
plt.ylabel("")
plt.grid()
plt.legend()
plt.show()
```

```python
bar_labels = ['Python', 'Java', 'C++', 'Go']
bar_values = [90, 75, 60, 30]
colors=["Purple","Green","Yellow","Red"]

sleep_hours = [4, 5, 6, 7, 8, 9, 10]
test_scores = [50, 55, 60, 70, 75, 78, 76]

x = np.linspace(0, 2*np.pi, 500)
y = np.sin(x)

pie_labels = ['Apple', 'Samsung', 'Xiaomi', 'Huawei']
market_share = [30, 25, 20, 25]
pie_colors = ['yellow', 'blue', 'green', 'red']

fig,axes = plt.subplots(nrows=2, ncols=2,figsize=(12, 14),dpi=100)
(ax1,ax2),(ax3,ax4) = axes

ax1.bar(bar_labels,bar_values,color=colors)
ax1.set_title("   ")
ax1.set_xlabel("")
ax1.set_ylabel("")
ax1.grid(axis="y",linestyle="-",alpha = 1)
ax2.scatter(sleep_hours,test_scores,color = "Blue",s = 100)
ax2.set_title("   ")
ax2.set_xlabel(" ")
ax2.set_ylabel("")
ax2.grid()
ax3.plot(x,y,color = "Black",label = "y = sin(x)")
ax3.set_title("y = sin(x)")
ax3.set_xlabel("X ")
ax3.set_ylabel("")
ax3.legend()
ax3.grid()
ax4.pie(market_share,labels=pie_labels,colors = pie_colors,autopct="%1.1f%%",startangle=180,explode=[0.1,0.0,0.1,0.0])
ax4.set_title("  ")
```
