# Заметка
В Python нельзя писать "3x" как в математике — он думает, что ты написал "три целых икс десятых" 😅
А надо писать с умножением явно.

```python
x = np.linspace(0,2*np.pi,500)
y1 = np.sin(x)
y2 = 2 * np.sin(x)
y3 = np.sin(x + np.pi / 2)

plt.plot(x,y1,color = "Green", label = "np.sin(x)")
plt.plot(x,y2,color = "Red", label = "2 * np.sin(x)")
plt.plot(x,y3,color = "Purple", label = "np.sin(x + np.pi / 2)")
plt.title("Сравнение синусоид")
plt.ylabel("Значения функций")
plt.xlabel("x (радианы)")
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace(0,2*np.pi,500)
y1 = np.sin(x) + np.cos(x)
y2 = np.sin(x) * np.cos(x)

plt.plot(x,y1,color = "Red", label = "np.sin(x) + np.cos(x)")
plt.plot(x,y2,color = "Green", label = "np.sin(x) * np.cos(x)")
plt.title("Сложение и умножение sin(x), cos(x)")
plt.xlabel("x (радианы)")
plt.ylabel("Значение")
plt.grid()
plt.legend()
plt.show()
```

# Заметка 
## Управление волнами Общая формула 
## y=A⋅sin(Bx+φ)
1. A <b>Что это?</b> = Амплитуда (высота волны)
<b>Как влияет?</b> = Чем больше A → тем выше/ниже волна
2. B <b>Что это?</b> = Частота (сколько волн в промежутке)
   <b>Как влияет?</b> = Увеличь B → волна "сжимается"
3. φ (фи) <b>Что это?</b> = Фаза (сдвиг волны)
   <b>Как влияет?</b> = Сдвигает волну влево/вправо (Значение в раиданах)

# Полярные координаты

## В обычных графиках ты рисуешь:
→ по X (влево/вправо)
↑ по Y (вверх/вниз)

## А в полярных координатах ты рисуешь так:

θ (тета) — угол (в радианах), то есть в какую сторону от центра
r — радиус, то есть насколько далеко от центра

```python
theta = np.linspace(0,2*np.pi,500)
r = np.sin(5*theta)
plt.polar(theta,r)
plt.title("Полярный график")
plt.show()
```

```python
theta = np.linspace(0,2*np.pi,500)
r= 1+np.sin(3*theta)

plt.polar(theta,r,color = "Red",label="r=1+sin(3θ)")
plt.legend()
plt.title("Цветок")
plt.show()
```

```python
theta = np.linspace(0,4*np.pi,500)
r = theta

plt.polar(theta,r,color = "Red", label = "r=θ")
plt.title("Спираль")
plt.legend()
plt.show()
```

# 🔸 1. Что такое фаза? Фаза — это сдвиг волны по горизонтали
# В формуле: y=A⋅sin(Bx+φ)

φ (фи) — это фаза
Если φ > 0 → волна сдвигается влево
Если φ < 0 → волна сдвигается вправо

```python
x = np.linspace(0,2*np.pi,500)
y1=np.sin(x)
y2=np.sin(x+np.pi/4)
y3=np.sin(x-np.pi/4)

plt.plot(x,y1,color="Green", label="np.sin(x)")
plt.plot(x,y2,color="Red", label="np.sin(x+np.pi/4)")
plt.plot(x,y3,color="Purple", label="np.sin(x-np.pi/4)")
plt.title("Сравнение фаз")
plt.xlabel("X Радианы")
plt.ylabel("ФАзы")
plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace(0,2*np.pi,500)
y1=np.sin(x)
y2=np.sin(3*x)
y3=np.sin(x)+0.5*np.sin(3 * x)

plt.plot(x,y1,color="Green", label="np.sin(x)")
plt.plot(x,y2,color="Red", label="np.sin(3x)")
plt.plot(x,y3,color="Purple", label="np.sin+0.5*np.sin(3x)")
plt.title("Сложение частот")
plt.xlabel("X Радианы")
plt.ylabel("Значения")
plt.grid()
plt.legend()
plt.show()
```

# Заметка
🔍 Параметры plt.pie:

values = Размеры долей (числа, не обязательно %)
labels = Подписи к каждой доле
colors = Цвета каждого сектора (как в bar)
autopct='%1.1f%%' = Автоматически подписывает проценты
startangle=90 = С какой стороны начинается круг
explode=[] = Смещение сектора (для выделения)

```python
labels = ['Apple', 'Samsung', 'Xiaomi', 'Huawei']
market_share = [30, 25, 20, 25]
colors = ["Yellow","Blue","Green","Red"]

plt.pie(market_share,labels=labels,colors = colors,autopct='%1.1f%%',startangle=90,explode=[0.1, 0, 0, 0])
plt.title("Доли на рынке смартфонов")
plt.show()
```
