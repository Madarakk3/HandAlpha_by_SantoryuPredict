# Заметка
🧠 1. Что такое логистическая регрессия?

Это модель классификации, которая предсказывает вероятность того, что объект относится к классу 1 (в нашем случае — "успешный").

Пример:
Вводим в модель:
пол = female
курс пройден
балл по чтению = 80
балл по математике = 72
Модель скажет:

вероятность успеха по письму = 0.84 → значит класс = 1 (успешный)

```python
df["reading success"] = (df["reading score"]>=70).astype(int)
X = df[["gender","lunch","test preparation course","math score","writing score"]]
Y = df["reading success"]

X_encoded = pd.get_dummies(X,drop_first = True)

X_train, X_test , Y_train, Y_test = train_test_split(
    X_encoded,Y,test_size = 0.3,random_state = 0
)
model = LogisticRegression(max_iter = 1000)
model.fit(X_train,Y_train)

y_pred = model.predict(X_test)
print(y_pred)
```

```python
accuracy = accuracy_score(Y_test,y_pred)
print(f"Точность предсказание",accuracy)

cm = confusion_matrix(Y_test,y_pred)
print("Confusion_Matrix:")
print(cm)

plt.figure(figsize=(6,5))
sns.heatmap(cm,annot = True,fmt = "d", cmap = "coolwarm")
plt.title("Матрица ошибок (Confusion Matrix)")
plt.xlabel("Предсказано")
plt.ylabel("Фактически")
plt.show()

print("Отчет по классификации")
print(classification_report(Y_test,y_pred))
# 1 (успех)	    143	✅ True Positive (TP)
# 0 (неуспех)	133	✅ True Negative (TN)
# 0 (неуспех)	16	❌ False Positive (FP) — модель сказала "успех", но это ошибка
# 1 (успех)	    8	❌ False Negative (FN) — модель не заметила успех
```

# Заметка
1️⃣ Precision (Точность)

Из всех, кого модель назвала "успешными", сколько действительно были успешными?

2️⃣ Recall (Полнота)

Из всех, кто действительно успешен, сколько модель правильно нашла?

3️⃣ F1-score — это баланс между Precision и Recall

Если precision и recall оба важны, F1 объединяет их в одно число.

```python
# Бинарная классификация 
df["writing success"] = (df["writing score"]>=70).astype(int)
X = df[["gender","lunch","test preparation course","math score","writing score"]]
Y = df["writing success"]

X_encoded = pd.get_dummies(X,drop_first = True)

X_train,X_test,Y_train,Y_test = train_test_split(
    X_encoded,Y,test_size = 0.3, random_state = 0
)

model = LogisticRegression(max_iter = 1000)
model.fit(X_train,Y_train)

y_pred = model.predict(X_test)

accuracy =  accuracy_score(Y_test,y_pred)
print("Точность предсказния", accuracy)

confusion = confusion_matrix(Y_test,y_pred)
print("Ошибки")
print( confusion)

plt.figure(figsize = (6,4))
sns.heatmap(confusion,annot = True,fmt = "d",cmap = "coolwarm")
plt.title("Confusion Matrix")
plt.xlabel("Предсказание")
plt.ylabel("Фактически")
plt.show()
# 1 (успех)	    152	✅ True Positive (TP)
# 0 (неуспех)	148	✅ True Negative (TN)
# 0 (неуспех)	0	❌ False Positive (FP) — модель сказала "успех", но это ошибка
# 1 (успех)	    0	❌ False Negative (FN) — модель не заметила успех

print("Отчет")
print(classification_report(Y_test,y_pred))

feature_name = X_encoded.columns
coefficient = model.coef_[0]

influence_df = pd.DataFrame({
    "Признаки":feature_name,
    "Вес(Влияние)":coefficient
})

influence_df = influence_df.sort_values(by = "Вес(Влияние)",ascending = True )
print(influence_df)
```

# Заметка
🔹.sort_values(...)
Метод .sort_values() используется, чтобы отсортировать строки в таблице
(по возрастанию или убыванию — ты сам выбираешь)

🔹 by="Вес (влияние)"
Ты указываешь, по какому столбцу сортировать
Здесь ты говоришь:

«Отсортируй DataFrame по значению столбца "Вес (влияние)"»
🔹 ascending=False
Это значит:

Сортировать по убыванию (от большого веса к меньшему)
Если бы ты написал ascending=True — это была бы сортировка по возрастанию

```python
# 1. Создание переменной
df["math level"] = pd.cut(
  df["math score"],
    bins = [0,60,80,100],
    labels = ["Низкий уровень","Средний уровень","Высокий уровень"],
    include_lowest=True # С помощью include_lowest можно расширить границы так как 100 значение является самым высоким и выдает Nan
)
# 2. Признаки и цель  
X = df[["gender","lunch","race/ethnicity","test preparation course","reading score","writing score"]]
Y = df["math level"]
# 3. One-hot encoding 
X_encoded = pd.get_dummies(X,drop_first = True)
# 4. Разделение 
X_train,X_test,Y_train,Y_test = train_test_split(
    X_encoded,Y,test_size = 0.3,random_state = 0
)
# 5. Обучение 
model = LogisticRegression(multi_class="multinomial", solver="lbfgs", max_iter=10000)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
# 6. Оценка
accuracy = accuracy_score(Y_test,y_pred)
confusion = confusion_matrix(Y_test,y_pred)
print("Точность предсказания",accuracy)
print("Ошибки",confusion)
print("Отчет",classification_report(Y_test,y_pred))
plt.figure(figsize=(6,4))
sns.heatmap(confusion,annot = True,fmt = "d",cmap = "coolwarm")
plt.title("Confusion Matrix")
plt.xlabel("Предсказание")
plt.ylabel("Фактически")
plt.show()
# 7. Интерпретация
for i, class_label in enumerate(model.classes_):
    print(f"Класс: {class_label}")
    for name,coef in zip(X_encoded,model.coef_[i]):
        print(f"{name}: {coef:3f}")

# 8. Визуализация 
for i,class_label in enumerate(model.classes_):
    plt.figure(figsize = (6,4))
    plt.barh(X_encoded.columns,model.coef_[i],color = 'skyblue')
    plt.title(f"Влияние признаков на класс: {class_label}")
    plt.xlabel("Коэффициент")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
