```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline 
from sklearn.linear_model import LogisticRegression 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import precision_score,recall_score,precision_recall_curve 
from sklearn.metrics import PrecisionRecallDisplay
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df["math success"] = (df["math score"]>=70).astype(int)
X = df[['gender', 'lunch', 'test preparation course', 'reading score', 'writing score']]
Y = df['math success']

X_encoded = pd.get_dummies(X,drop_first=True)
print(X_encoded.head())
```

# –ó–∞–º–µ—Ç–∫–∞
üß† 1. –ß—Ç–æ —Ç–∞–∫–æ–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è?

–≠—Ç–æ –º–æ–¥–µ–ª—å –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–±—ä–µ–∫—Ç –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∫ –∫–ª–∞—Å—Å—É 1 (–≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ ‚Äî "—É—Å–ø–µ—à–Ω—ã–π").

–ü—Ä–∏–º–µ—Ä:
–í–≤–æ–¥–∏–º –≤ –º–æ–¥–µ–ª—å:
–ø–æ–ª = female
–∫—É—Ä—Å –ø—Ä–æ–π–¥–µ–Ω
–±–∞–ª–ª –ø–æ —á—Ç–µ–Ω–∏—é = 80
–±–∞–ª–ª –ø–æ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ = 72
–ú–æ–¥–µ–ª—å —Å–∫–∞–∂–µ—Ç:

–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É—Å–ø–µ—Ö–∞ –ø–æ –ø–∏—Å—å–º—É = 0.84 ‚Üí –∑–Ω–∞—á–∏—Ç –∫–ª–∞—Å—Å = 1 (—É—Å–ø–µ—à–Ω—ã–π)

```python
df["reading success"] = (df["reading score"]>=70).astype(int)
X = df[["gender","lunch","test preparation course","math score","writing score"]]
Y = df["reading success"]

X_encoded = pd.get_dummies(X,drop_first = True)

X_train, X_test , Y_train, Y_test = train_test_split(
    X_encoded,Y,test_size = 0.3,random_state = 0
)
model = LogisticRegression(max_iter = 1000)
model.fit(X_train,Y_train)

y_pred = model.predict(X_test)
print(y_pred)
```

```python
accuracy = accuracy_score(Y_test,y_pred)
print(f"–¢–æ—á–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ",accuracy)

cm = confusion_matrix(Y_test,y_pred)
print("Confusion_Matrix:")
print(cm)

plt.figure(figsize=(6,5))
sns.heatmap(cm,annot = True,fmt = "d", cmap = "coolwarm")
plt.title("–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫ (Confusion Matrix)")
plt.xlabel("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–æ")
plt.ylabel("–§–∞–∫—Ç–∏—á–µ—Å–∫–∏")
plt.show()

print("–û—Ç—á–µ—Ç –ø–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏")
print(classification_report(Y_test,y_pred))
# 1 (—É—Å–ø–µ—Ö)	    143	‚úÖ True Positive (TP)
# 0 (–Ω–µ—É—Å–ø–µ—Ö)	133	‚úÖ True Negative (TN)
# 0 (–Ω–µ—É—Å–ø–µ—Ö)	16	‚ùå False Positive (FP) ‚Äî –º–æ–¥–µ–ª—å —Å–∫–∞–∑–∞–ª–∞ "—É—Å–ø–µ—Ö", –Ω–æ —ç—Ç–æ –æ—à–∏–±–∫–∞
# 1 (—É—Å–ø–µ—Ö)	    8	‚ùå False Negative (FN) ‚Äî –º–æ–¥–µ–ª—å –Ω–µ –∑–∞–º–µ—Ç–∏–ª–∞ —É—Å–ø–µ—Ö
```

# –ó–∞–º–µ—Ç–∫–∞
1Ô∏è‚É£ Precision (–¢–æ—á–Ω–æ—Å—Ç—å)

–ò–∑ –≤—Å–µ—Ö, –∫–æ–≥–æ –º–æ–¥–µ–ª—å –Ω–∞–∑–≤–∞–ª–∞ "—É—Å–ø–µ—à–Ω—ã–º–∏", —Å–∫–æ–ª—å–∫–æ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±—ã–ª–∏ —É—Å–ø–µ—à–Ω—ã–º–∏?

2Ô∏è‚É£ Recall (–ü–æ–ª–Ω–æ—Ç–∞)

–ò–∑ –≤—Å–µ—Ö, –∫—Ç–æ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —É—Å–ø–µ—à–µ–Ω, —Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ –Ω–∞—à–ª–∞?

3Ô∏è‚É£ F1-score ‚Äî —ç—Ç–æ –±–∞–ª–∞–Ω—Å –º–µ–∂–¥—É Precision –∏ Recall

–ï—Å–ª–∏ precision –∏ recall –æ–±–∞ –≤–∞–∂–Ω—ã, F1 –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç –∏—Ö –≤ –æ–¥–Ω–æ —á–∏—Å–ª–æ.

```python
# –ë–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è 
df["writing success"] = (df["writing score"]>=70).astype(int)
X = df[["gender","lunch","test preparation course","math score","writing score"]]
Y = df["writing success"]

X_encoded = pd.get_dummies(X,drop_first = True)

X_train,X_test,Y_train,Y_test = train_test_split(
    X_encoded,Y,test_size = 0.3, random_state = 0
)

model = LogisticRegression(max_iter = 1000)
model.fit(X_train,Y_train)

y_pred = model.predict(X_test)

accuracy =  accuracy_score(Y_test,y_pred)
print("–¢–æ—á–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–Ω–∏—è", accuracy)

confusion = confusion_matrix(Y_test,y_pred)
print("–û—à–∏–±–∫–∏")
print( confusion)

plt.figure(figsize = (6,4))
sns.heatmap(confusion,annot = True,fmt = "d",cmap = "coolwarm")
plt.title("Confusion Matrix")
plt.xlabel("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ")
plt.ylabel("–§–∞–∫—Ç–∏—á–µ—Å–∫–∏")
plt.show()
# 1 (—É—Å–ø–µ—Ö)	    152	‚úÖ True Positive (TP)
# 0 (–Ω–µ—É—Å–ø–µ—Ö)	148	‚úÖ True Negative (TN)
# 0 (–Ω–µ—É—Å–ø–µ—Ö)	0	‚ùå False Positive (FP) ‚Äî –º–æ–¥–µ–ª—å —Å–∫–∞–∑–∞–ª–∞ "—É—Å–ø–µ—Ö", –Ω–æ —ç—Ç–æ –æ—à–∏–±–∫–∞
# 1 (—É—Å–ø–µ—Ö)	    0	‚ùå False Negative (FN) ‚Äî –º–æ–¥–µ–ª—å –Ω–µ –∑–∞–º–µ—Ç–∏–ª–∞ —É—Å–ø–µ—Ö

print("–û—Ç—á–µ—Ç")
print(classification_report(Y_test,y_pred))

feature_name = X_encoded.columns
coefficient = model.coef_[0]

influence_df = pd.DataFrame({
    "–ü—Ä–∏–∑–Ω–∞–∫–∏":feature_name,
    "–í–µ—Å(–í–ª–∏—è–Ω–∏–µ)":coefficient
})

influence_df = influence_df.sort_values(by = "–í–µ—Å(–í–ª–∏—è–Ω–∏–µ)",ascending = True )
print(influence_df)
```

# –ó–∞–º–µ—Ç–∫–∞
üîπ.sort_values(...)
–ú–µ—Ç–æ–¥ .sort_values() –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è, —á—Ç–æ–±—ã –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å —Å—Ç—Ä–æ–∫–∏ –≤ —Ç–∞–±–ª–∏—Ü–µ
(–ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é –∏–ª–∏ —É–±—ã–≤–∞–Ω–∏—é ‚Äî —Ç—ã —Å–∞–º –≤—ã–±–∏—Ä–∞–µ—à—å)

üîπ by="–í–µ—Å (–≤–ª–∏—è–Ω–∏–µ)"
–¢—ã —É–∫–∞–∑—ã–≤–∞–µ—à—å, –ø–æ –∫–∞–∫–æ–º—É —Å—Ç–æ–ª–±—Ü—É —Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å
–ó–¥–µ—Å—å —Ç—ã –≥–æ–≤–æ—Ä–∏—à—å:

¬´–û—Ç—Å–æ—Ä—Ç–∏—Ä—É–π DataFrame –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é —Å—Ç–æ–ª–±—Ü–∞ "–í–µ—Å (–≤–ª–∏—è–Ω–∏–µ)"¬ª
üîπ ascending=False
–≠—Ç–æ –∑–Ω–∞—á–∏—Ç:

–°–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–æ —É–±—ã–≤–∞–Ω–∏—é (–æ—Ç –±–æ–ª—å—à–æ–≥–æ –≤–µ—Å–∞ –∫ –º–µ–Ω—å—à–µ–º—É)
–ï—Å–ª–∏ –±—ã —Ç—ã –Ω–∞–ø–∏—Å–∞–ª ascending=True ‚Äî —ç—Ç–æ –±—ã–ª–∞ –±—ã —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é

```python
df.head(10)
```

```python
# 1. –°–æ–∑–¥–∞–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
df["math level"] = pd.cut(
  df["math score"],
    bins = [0,60,80,100],
    labels = ["–ù–∏–∑–∫–∏–π —É—Ä–æ–≤–µ–Ω—å","–°—Ä–µ–¥–Ω–∏–π —É—Ä–æ–≤–µ–Ω—å","–í—ã—Å–æ–∫–∏–π —É—Ä–æ–≤–µ–Ω—å"],
    include_lowest=True # –° –ø–æ–º–æ—â—å—é include_lowest –º–æ–∂–Ω–æ —Ä–∞—Å—à–∏—Ä–∏—Ç—å –≥—Ä–∞–Ω–∏—Ü—ã —Ç–∞–∫ –∫–∞–∫ 100 –∑–Ω–∞—á–µ–Ω–∏–µ —è–≤–ª—è–µ—Ç—Å—è —Å–∞–º—ã–º –≤—ã—Å–æ–∫–∏–º –∏ –≤—ã–¥–∞–µ—Ç Nan
)
# 2. –ü—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª—å  
X = df[["gender","lunch","race/ethnicity","test preparation course","reading score","writing score"]]
Y = df["math level"]
# 3. One-hot encoding 
X_encoded = pd.get_dummies(X,drop_first = True)
# 4. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ 
X_train,X_test,Y_train,Y_test = train_test_split(
    X_encoded,Y,test_size = 0.3,random_state = 0
)
# 5. –û–±—É—á–µ–Ω–∏–µ 
model = LogisticRegression(multi_class="multinomial", solver="lbfgs", max_iter=10000)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
# 6. –û—Ü–µ–Ω–∫–∞
accuracy = accuracy_score(Y_test,y_pred)
confusion = confusion_matrix(Y_test,y_pred)
print("–¢–æ—á–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è",accuracy)
print("–û—à–∏–±–∫–∏",confusion)
print("–û—Ç—á–µ—Ç",classification_report(Y_test,y_pred))
plt.figure(figsize=(6,4))
sns.heatmap(confusion,annot = True,fmt = "d",cmap = "coolwarm")
plt.title("Confusion Matrix")
plt.xlabel("–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ")
plt.ylabel("–§–∞–∫—Ç–∏—á–µ—Å–∫–∏")
plt.show()
# 7. –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è
for i, class_label in enumerate(model.classes_):
    print(f"–ö–ª–∞—Å—Å: {class_label}")
    for name,coef in zip(X_encoded,model.coef_[i]):
        print(f"{name}: {coef:3f}")

# 8. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è 
for i,class_label in enumerate(model.classes_):
    plt.figure(figsize = (6,4))
    plt.barh(X_encoded.columns,model.coef_[i],color = 'skyblue')
    plt.title(f"–í–ª–∏—è–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞ –∫–ª–∞—Å—Å: {class_label}")
    plt.xlabel("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
