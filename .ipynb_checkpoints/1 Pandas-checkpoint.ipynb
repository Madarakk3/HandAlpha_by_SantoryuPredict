### Заметка
#### Когда нужны скобки?
- В логических операциях (|, &, ~), чтобы задать порядок выполнения:
- В вызовах методов, когда передаём аргументы

(df["Age"].isin([10, 5, 15])  # Метод .isin() требует скобки для списка)

- Если метод без аргументов, скобки можно опустить:
#### Когда скобки НЕ нужны?
- В .iloc[] и .loc[], если индексация простая:
- В .isin() и других методах, если они уже возвращают булевый массив:
#### Главное правило:
- Если есть логические операции (|, &, ~) → нужны скобки.
- Если вызываешь метод с аргументами → нужны скобки.
- Если просто индексация (iloc, loc) → не нужны скобки.

 # Сортировка данных

### Заметка
- Индексация — это способ обращения к данным в DataFrame или Series. Она позволяет выбирать отдельные значения, строки, столбцы или их сочетания.
- Пример: *df.loc[], df.loc[5, "Age"], df.iloc[5, 2], df["col"], df["Age"]*

```python
df.sort_values("Age").head(5) #Сортировка от самого маленького значения до самого большого
```

```python
df.sort_values(["Age","Name"], ascending=[False,True]).head(5) #Сортировка от самого большого значения до самого маленького
```

# Объединение датафреймов с помощью concat и merge

```python
df2=df.copy(deep=True)
```

```python
cdf2=pd.concat([df,df2])
```

```python
cdf2.shape
```

```python
cdf2
```

```python
mtf=pd.DataFrame(index=df.index)
mtf["PassengerId"]=df["PassengerId"]
mtf["evenid"]=mtf["PassengerId"].apply(lambda x: x % 2 == 0)
```

```python
mtf
```

```python
pd.merge(df,mtf,how="inner")
```
